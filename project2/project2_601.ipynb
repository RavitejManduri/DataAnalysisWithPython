# Hypothesis 1

## Higher the population higher the number of zoos in the country?

```python
import numpy as np
# Groupby country
m=zoo.groupby(['country']).sum()
barWidth = 0.25
r1 = np.arange(5)
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
plt.bar(r1, m['humans']/1000, color='#7f6d5f', width=barWidth, edgecolor='white', label='var1')
plt.bar(r2, m['animals'], color='#557f2d', width=barWidth, edgecolor='white', label='var2')

plt.xlabel('countries', fontweight='bold')
plt.ylabel('population in normalised form')
plt.xticks([r + barWidth for r in range(5)], ['Brazil', 'Canada', 'China', 'India', 'United States'])
plt.title('Population of animals and humans in each country')
```

In the above graph humans population is divided by 1000 so that the comparisio graph would be more clearer.

```python
m
```

## Validation of hypothesis 1

Surprisingly my hypothesis 1 is perfectly true for the data.

# Hypothesis 3

## China has most no of zoos

```python
_=zoo['country'].value_counts().plot(kind="bar")
_=plt.ylabel('Zoo count',fontsize=12)
_=plt.xlabel('Country',fontsize=12)
_=plt.title('Zoo count in each Country')
```

## Validation Of Hypothesis 3

From the graph we can see that china is dominating in number of zoo in its country. From this we can say that our hypothesis 3 of China has most no of zoos is absolutely correct.
