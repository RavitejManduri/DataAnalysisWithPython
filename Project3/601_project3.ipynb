Finding the no of Restaurants in each state using the below for loop and storing it in states dictionary that we created previously

This loop just increases the counter in a state if it is found in restaurant dataset

```python
for index, row in restaurant.iterrows():
    states[row['province']]=states[row['province']]+1
    #print(states[row['province']])
states
```

Now adding this states and restaurant count data to the population dataframe(pd) that we created previously

```python

pd['Restaurant_count']=""
for index,row in pd.iterrows():
    if row['State'] in states:
        pd.loc[index, "Restaurant_count"]=states[row['State']]
    else:
        pd.loc[index, "Restaurant_count"]=0
```

Finding the avg_obesity of each state and adding it to the pd dataframe

```python
obese=population.groupby(['LocationAbbr']).mean().apply(list).to_dict()
```

```python
pd['Avg_obesity']=""
for index,row in pd.iterrows():
    if row['State'] in states:
        pd.loc[index, 'Avg_obesity']=obesity[row['State']]
    else:
        pd.loc[index, 'Avg_obesity']=0
```

```python
df=df.reset_index()
```

```python
df.columns=["name","count"]
```

```python
df
```

```python
df.groupby("name").sum().sort_values('count', ascending=False)
```

```python
import matplotlib.pyplot as plt
ax = df.groupby("name").sum().sort_values('count', ascending=False).plot(kind='bar', title ="Count of Restaurants", figsize=(20, 8), legend=True, fontsize=12)
ax.set_xlabel("Restaurant", fontsize=12)
ax.set_ylabel("Count Of Restaurant", fontsize=12)
plt.show()
```

```python

bx = df.groupby("name").sum().sort_values('count', ascending=False).plot( title ="Count of Restaurants", figsize=(20, 8), legend=True, fontsize=12)
bx.set_xlabel("Restaurant", fontsize=12)
bx.set_ylabel("Count Of Restaurant", fontsize=12)
plt.show()
```
