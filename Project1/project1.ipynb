# Cleaning:
Now I am replacing "dollar" and ',' with none in tuition and fees column and converting it into an integer field for further processing so that $41,897 becomes 41897.

```python
ranking['Tuition and fees'] = ranking['Tuition and fees'].str.replace('$','').str.replace(',','').astype('int64')
```

```python
ranking
```

# filling missing values
Few of the fields in 'In-state' column are missing so I filled them with zeros for now.

```python
ranking['In-state'] = ranking['In-state'].fillna(0)
```

Now I am replacing "dollar" and ',' with none in In-state column as well and converting it into an integer field for further processing so that $41,897 becomes 41897.

```python
ranking['In-state'] = ranking['In-state'].astype(str).str.replace('$','').str.replace(',','').astype('int64')
```

```python
ranking
```

Removing ',' from the Undergrad Enrollment column and replacing with none and finally converting into an integer

```python
ranking['Undergrad Enrollment'] = ranking['Undergrad Enrollment'].astype(str).str.replace(',','').astype('int64')
```

# splitting one column into two
'Location' column has place and state now I am splitting it into two separate fields
the only string that splits a place and a state here is ', ' so I am replacing it with a dollar as a delimiter

```python
ranking['Location'] = ranking['Location'].str.replace(', ','$')
```

https://www.geeksforgeeks.org/python-pandas-split-strings-into-two-list-columns-using-str-split/
This link has the code to split the column as shown below

```python
#ranking
new = ranking["Location"].str.split("$", n = 1, expand = True) 
ranking["Place"]= new[0]
ranking["state"]= new[1]
ranking.drop(columns =["Location"], inplace = True)
```

```python
ranking
```

# Filling the zeros (in In-state field) with their exact values
here I am dividing the 'Tuition and fees' with the respective known 'In-state' field and getting an avg of all these values.

```python
sum=0
count=0
for i in range(ranking.shape[0]):
    if(ranking['In-state'][i]>0):
        count=count+1
        sum=sum+(ranking['Tuition and fees'][i]/ranking['In-state'][i])
avg=sum/count
avg
```

```python
ranking
```

Finally I am dividing the Tuition and fees with the avg we found previously and filling the missing values in In-state field.

I am doing this because basically Tuition and fees and In-state-fees are related to each other.

```python
for i in range(ranking.shape[0]):
    if(ranking['In-state'][i]==0):
        x=ranking['Tuition and fees'][i]
        ranking['In-state'][i]=x/avg
        
```
