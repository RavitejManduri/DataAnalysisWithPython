```python
import pandas
import matplotlib.pyplot as plt
```

```python
from matplotlib.pyplot import figure
```

The data link which I used is https://data.world/baltimore/k5ry-ef3g/workspace/file?filename=restaurants-1.csv

```python
restaurants=pandas.read_csv('restaurants-1.csv')
restaurants.dtypes
```

Cleaned the data by deleting unwanted columns.

As there is no information in the last three columns I deleted those columns. https://stackoverflow.com/questions/13411544/delete-column-from-pandas-dataframe referred this link to delete the columns.

```python
del restaurants['2010 Census Neighborhoods']
```

```python
del restaurants['2010 Census Wards Precincts']
```

```python
del restaurants['Zip Codes']
```

```python
restaurants.shape
```

Referred the link https://www.geeksforgeeks.org/working-with-missing-data-in-pandas/ to check whether there are any null values or not

```python
restaurants.isnull().values.any()
```

Here in the below chart I visualized the number of restaurants in each region.

```python
_=restaurants['policeDistrict'].value_counts().plot(kind="bar")
_=plt.ylabel('Restaurant Count',fontsize=15)
_=plt.xlabel('region',fontsize=15)
```

The below scatter plot represents the number(council district) in each region.The dark colored plots represents the higher density and the light colored density represents the lesser density.

```python
x=restaurants['councilDistrict']
y=restaurants['policeDistrict']
plt.scatter(x, y,alpha=0.5)
plt.xlabel('Area')
plt.ylabel('Region')
```

The below bar chart shows the number of restaurants in different places(Zipcode)

```python
plt.plot(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
_=restaurants['zipCode'].value_counts().plot(kind="bar")
_=plt.ylabel('Total restaurants',fontsize=11)
```
